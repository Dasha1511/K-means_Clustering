```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X['Cluster'] = cluster_labels
```

```python
X
```

```python
plt.figure(figsize=(12,6), dpi = 200)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind = 'bar')
```

```python
ssd = []

for k in range(2, 10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_) #сумма квадратов расстояний от точек до центров кластеров
```

```python
ssd
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(2,10), ssd, 'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()  #разница междц текущем и предыдущем элементом
```

```python
pd.DataFrame(data=zip(range(2,10), pd.Series(ssd).diff()), columns=['K', 'SSD diff'])
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouettes = []

for k in range(2, 10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    silhouettes.append(silhouette_score(scaled_X, model.labels_)) #
```

```python
silhouettes
```
