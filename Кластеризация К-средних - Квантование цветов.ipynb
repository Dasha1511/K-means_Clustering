```python
# (H, W, C) => (H*W, C)
```

```python
(h, w, c) = image_as_array.shape
```

```python
h
```

```python
w
```

```python
c
```

```python
image_as_array2d = image_as_array.reshape(h*w, c)
```

```python
image_as_array2d
```

```python
len(image_as_array.shape)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=100)
```

```python
labels = model.fit_predict(image_as_array2d) # разделение на кластеры
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int) # 6 цветов
```

```python
rgb_codes
```

```python
quantized_image = np.reshape(rgb_codes[labels], (h, w, c))
```

```python
image_as_array
```

```python
quantized_image
```

```python
plt.figure(figsize=(8, 4), dpi = 200)
plt.imshow(quantized_image)
```
